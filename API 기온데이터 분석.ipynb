### 매년 내생일 최고 최저 기온 시각화 하기

```python

f= open('seoul.csv')
data= csv.reader(f)
header= next(data)

max_date=[]
high =[]
low=[]


for row in data:
    if row[-1]== '':
        row[-1]= -999
    if row[0].split('-')[0] >='1998':
        if row[0].split('-')[1]=='07'and row[0].split('-')[2]=='11':
            max_date.append(row[0].split('-')[0])
            high.append(float(row[-1]))
            low.append(float(row[-2]))



f.close()
print(max_date)
print(low)
print(high)


plt.figure(figsize=(8, 8))
plt.plot(max_date,high,'hotpink')
plt.plot(low,'skyblue')
plt.xticks(rotation=50)

plt.rc('font', family='Malgun Gothic')
plt.title('내 생일에 기후 변화')
plt.show()
```

### 1월8월 최고기온을 히스토그램으로 표현하기

```python
f= open('seoul.csv')
data= csv.reader(f)
header= next(data)
result =[]
au=[]
for row in data:
    if row[-1] !='':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))
        elif row[0].split('-')[1] == '01':
            au.append(float(row[-1]))

plt.hist(result,bins=100)
plt.hist(au,bins=100)
f.close()
```

### boxplot 상자그림으로 표현하기

```python
1~12월 모든 최고기온을 하나의 리스트에 넣어서 박스그림으로 표현
```

```python
f= open('seoul.csv')
data= csv.reader(f)
header= next(data)
month=[[],[],[],[],[],[],[],[],[],[],[],[]]


for row in data:
    if row[-1] !='':
        # 리스트 인덱스와 날짜가 동일하면 append시킨다
        #1월 -1 하면 1월은 0인덱스
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))

print(month[0])
plt.boxplot(month)
f.close()
```
